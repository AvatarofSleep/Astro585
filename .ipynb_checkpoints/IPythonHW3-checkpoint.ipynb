```julia
function write_down(x::Array,filename::String)
    #Write an array to file  
    #INPUTS: filename -  file to be made, x - Array to be written
    #OUTPUTS: a file named "filename" containing a binary array
    f = open(filename,"w")
    write(f,x)
    close(f)
end

function read_up(filename::String,size)
    #Reads an array from a file  
    #INPUTS: filename -  file to be read, size - size of the array to be read. Can be less than the array in the file, 
    #cannot be larger than array
    #OUTPUTS: an array of size "size"
    f = open(filename,"r")
    z = read(f,Array(Float64,size))
    close(f)
end
    
```

# 3.E using HDF5

Functions made to test read write speeds

```julia
function write_HDF5(filename::String,title::String,x::Array)
    
    matrix = reshape(x, length(x),1)
    h5write(filename,title,matrix)
end

function read_HDF5(filename::String,title::String)
    outs = h5read(filename,title)
end

function read_HDF5_spec(filename::String,title::String,spot)
    val = h5read(filename,title,(spot:spot,)) #in order to read, you need the (m:m,) setup to tell it you are indexing
    return(val)
end

```

```julia
arra = linspace(1,100,(1024^2))

using HDF5, JLD
println(@elapsed(write_HDF5("hdfarray.jld","box",arra)))
println(@elapsed(read_HDF5("hdfarray.jld","box")))
println(@elapsed(read_HDF5_spec("hdfarray.jld","box",(23,1))))
```

For a 1028-element array, it takes 2.5797ms to write, 1.7781ms to read, and 1.8409ms to read a specific element. 

For a 1028^element array, it takes 11.00ms to write, 12.28ms to read, and 1.914 ms to read a specific element.

```julia
function elem_csv(filename::String,element)
    #Reads a CSV file and returns the chosen element of the array 
    #INPUTS: filename - filename of a CSV file, element - element to be returned
    #OUTPUTS: value of the element of the read array
    r = readcsv(filename)
    print(r[element])
end

function elem_bin(filename::String,size)
    #Reads a file and returns the chosen element of the array 
    #INPUTS: filename - filename of a binarr file conatning an array, size - element of the array to be returned
    #OUTPUTS: value of the element of the read array
    f = open(filename,"r")
    g = read(f,Array(Float64,size))
    close(f)
    r = g[size]
    
    print(r)
    return(r)
end
```

```julia
print(@elapsed(elem_bin("counted.txt",1024,500)))
#print(@elapsed(elem_csv("block.csv",500)))
```

to find an element midway through the array in a binary array it took 2.872ms and in a csv it took 6.842ms

It appears well suited to quick save and recall.

# 3.D

Variance checker. Kicks out numbers that aren't finite (NaN, inf). Returns variance of all good numbers.

```julia
function var_online_with_check(y::Array)
    #one pass variance calculation from Homework 1
    #INPUTS:
    #a 1 dimensional array of numbers
    #OUTPUTS:
    #variance of the input array
    #The major change from the standard var_online is that this will search for non finite numbers and toss them with a warning. 
  n = length(y);
  sum1 = zero(y[1]);
  mean = zero(y[1]);
  M2 = zero(y[1]);
    z=0
  for i in 1:n
        if isfinite(y[i])
	       diff_by_i = (y[i]-mean)/i;
            mean = mean +diff_by_i;
	        M2 = M2 + (i-1)*diff_by_i*diff_by_i+(y[i]-mean)*(y[i]-mean)
            else
            z = z+1
            println("value",i,"is no good. Ignoring")
        end
  end;  
  variance = M2/(n-1-z);
    return(variance)
end

```

```julia
using Base.Test
arr = randn(100)
red = var_online_with_check(arr)
blue = var(arr)
@test_approx_eq_eps(red,blue,.0000001)
```

# 3.E

The quick regression test is to see if the returned value is still the variance that is measured using var(x). So far so good.

```julia
function var_online(y::Array)
  n = length(y);
  sum1 = zero(y[1]);
  mean = zero(y[1]);
  M2 = zero(y[1]);
  for i in 1:n
	  diff_by_i = (y[i]-mean)/i;
	  mean = mean +diff_by_i;
	  M2 = M2 + (i-1)*diff_by_i*diff_by_i+(y[i]-mean)*(y[i]-mean); 
  end;  
  variance = M2/(n-1);
end
```

Once both functions were included in the cache, they had the exact same processing time. This is partially likely due to the lack of NaN and inf. But it effectively costs nothing to toss out bad numbers. if they are rarely occuring.
