# 1.b for-loop matrix vector

```julia
function MatVecFor(A::Array{Float64,2},b::Array{Float64,1})
    
    @assert size(A,2) == length(b)
    
    c = zeros(length(b))
    for i in 1:size(A,1)
        for j in 1:length(b)
            c[i] = c[i] + A[i,j]*b[j]
        end
    end
    
    return c
end
```

```julia
MatVecFor(A,b)
```

Returns same value!

```julia
function MatVecFor2(A::Array{Float64,2},b::Array{Float64,1})
    
    @assert size(A,2) == length(b)
    
    c = zeros(size(A,2),size(A,1))
    for i in 1:length(b)
        for j in 1:size(A,1)
            c[j,i] = A[j,i]*b[i]
        end
    end
    c = sum(c,2)
    return c
end
```

```julia
d = MatVecFor2(A,b)
```

Again the result is the same!

```julia
##from last HW ##
using Devectorize

function trap_devec_pdf(min, steps=10000)
    #inputs - func - a single variable function to be integrated min- integration minimum max-integration maximum
    #steps, optional number of steps the integral is broken into. preset to 10^4
    #outputs result - trapezoid integral of function func 
        
    stepsize = abs(2*min)/(steps-1)
    range = linspace(-min,min,steps)    
    minval = exp(-0.5*min^2)/sqrt(2*pi)
    maxval = exp(-0.5*min^2)/sqrt(2*pi)
    @devec values = 2.*exp(-0.5.*range.^2)./sqrt(2.*pi)
    
    integral = (sum(values)-minval-maxval)*stepsize/2
    
    return integral
end
```

```julia
lookup_table = make_table_linear(x->trap_devec_pdf(x),-10.,10.,1000)
```

```julia
x = linspace(-10,10,1000)
@time lookup(lookup_table,x)
@time trap_devec_pdf(10)
```

Here we get something where the lookup table is on a closer time with the original function.
